# Matplotlib Examples

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x **2
```

```python
x
```

```python
y
```

```python
# Functional Method
```

```python
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1,2,1) 
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#Object Oriented
fig =plt.figure()

# blank canvas
axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)

axes2.plot(y,x)

axes2.set_title('Small')
axes1.set_title('Big')
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[0].set_title('First')

axes[1].plot(y,x)
axes[1].set_title('Second')

plt.tight_layout()
```

## Figure Size and DPI

```python
# allows to control aspects
```

```python
fig = plt.figure(figsize =(3,2))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='x sqr')
ax.plot(x,x**3, label = 'x cubed')

ax.legend(loc = 10)

# loc will move legend around, 0 for best. 
# Alt use a tuple to choose your location
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize =(7,2))

axes[0].plot(x,y)

axes[1].plot(y,x)
```

```python
fig.savefig('myplot.png',dpi=200)
```

## Plot Appearance

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='green', lw = 4,alpha = 1,
        linestyle='--',marker='o',ms = 10,mfc = 'red') 
# or RGB Hex Code '#----'
# mfc = marker face colorhb
```
